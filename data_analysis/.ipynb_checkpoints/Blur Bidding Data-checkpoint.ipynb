```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as nx
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv(r'C:\Users\Connor\Desktop\NFT Data Analysis\blur_bidding2.csv')
df
```

```python
df.info()
```

```python
# See if the data has any missing values
df_missing = df.isnull().sum()
df_missing
```

```python
# Get rid of time 
df['Date'] = pd.to_datetime(df['Date']).dt.date
df['Date']
```

```python
code_counts = df["Date"].value_counts()
code_counts
```

```python
# Obtain top 10 address bidding
top_10 = df.groupby('From')['Value (ETH)'].sum().sort_values(ascending=False)[:10]

# Print the top 10 addresses and their total bid value
print(top_10)
```

I will rename any accounts who have a ENS name in their wallet. This is so we can identify easier rather than just an address

```python
top_10 = top_10.rename({'0xed2ab4948ba6a909a7751dec4f34f303eb8c7236':'franklinisbored.eth',
                        '0x1b0aa2c9a4abff338b755a1ed2df975ec1fd1064':'bluri0.eth',
                        '0xa4fb0e1f6ad0a87a3ec2427ca3a944c42d19d1ca':'jahoda.eth',})
top_10
```

# This bar chart finds the top 10 bidders on the blur marketplace in the last 24hrs(or however long of a duration your data is, mine is 24hr)

```python
# Create pastel color palette
flare_palette = sns.cubehelix_palette(10,gamma=.5)[::-1]

# Create bar plot 
plt.figure(figsize=(10,5))
ax = sns.barplot(x=top_10.index, y=top_10.values, color=flare_palette[1], palette=flare_palette)

# Add value labels to bars
for i, v in enumerate(top_10.values):
    ax.text(i, v+10, str(round(v,2)), horizontalalignment='center', fontsize=10)

# Set plot title and axis labels
plt.title("Top 10 Blur Bidders by Bid Value from 2/25", fontsize=16)
plt.xlabel("Bidders", fontsize=14)
plt.ylabel("Bid Value (ETH)", fontsize=14)

# Rotate x-axis labels for readability
plt.xticks(rotation=80)

# Show plot
plt.show()
```

Creating vairables of values between certain numbers

```python
bids_below_1eth = df[df['Value (ETH)'] < 1]['Value (ETH)'].count()
bids_below_1eth
```

```python
bids_below_10eth = df[(df['Value (ETH)'] > 1) & (df['Value (ETH)'] < 10)]['Value (ETH)'].count()
bids_below_10eth
```

```python
bids_below_20eth = df[(df['Value (ETH)'] > 10) & (df['Value (ETH)'] < 20)]['Value (ETH)'].count()
bids_below_20eth
```

```python
bids_below_50eth = df[(df['Value (ETH)'] > 20) & (df['Value (ETH)'] < 50)]['Value (ETH)'].count()
bids_below_50eth
```

```python
bids_above_50eth = df[df['Value (ETH)'] > 50]['Value (ETH)'].count()
bids_above_50eth
```

```python
values = [bids_below_1eth, bids_below_10eth, bids_below_20eth, 
          bids_below_50eth, bids_above_50eth]
labels = ['< 1 ETH', '< 10 ETH', '< 20 ETH', '< 50 ETH',
          '> 50 ETH']
```

```python
# Calculate percentage values
total = sum(values)
percentages = [(value/total)*100 for value in values]

# Create bar plot 
plt.figure(figsize=(10,5))
sns.barplot(x=percentages, y=labels, color=flare_palette[1], palette=flare_palette)

# Add percentage labels on top of each bar
for i, val in enumerate(percentages):
    plt.text(val, i, f'{val:.2f}%', color='black', ha='left', va='center')

    
plt.xlabel('Percentage of Bids')
plt.ylabel('Bid Amount Range')
plt.title('Percentage of Bids by ETH Value')
plt.show()
```
